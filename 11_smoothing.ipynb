```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

```python
img=cv2.imread("data/filter.jpg",1)
img1=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img)
plt.show()
```

```python

kernal=np.ones((5,5),np.float32)/25
dst=cv2.filter2D(img1,-1,kernal)
titles=["image","2dconv"]
imgs=[img1,dst]
for i in range(2):
    plt.subplot(1,2,i+1),plt.imshow(imgs[i],'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
plt.show()
```

Yes, essentially, both convolutions and the filter2D function serve the same purpose: applying a kernel to an input signal or image to perform a convolution operation

<li>low pass filter helps in removing noise,blurring</li>
<li>high pass filter helps in finding edges </li>

<b>Bluring</b><br>
to blur , convolve with lpf

```python

kernal=np.ones((5,5),np.float32)/25
dst=cv2.filter2D(img1,-1,kernal)
#This means you can use it for various operations like sharpening, edge detection, or custom filtering.
blur=cv2.blur(img1,(5,5))
#is specifically for uniform blurring using a simple averaging filter (box filter).



titles=["image","2dconv","blur"]
imgs=[img1,dst,blur]
for i in range(3):
    plt.subplot(1,3,i+1),plt.imshow(imgs[i],'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
plt.show()
```

### gaussian filter


#### designed specifically for removng high frequency noise from image

```python
img=cv2.imread("data/halftone.jpg",1)
img1=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

kernal=np.ones((5,5),np.float32)/25
dst=cv2.filter2D(img1,-1,kernal)

blur=cv2.blur(img1,(5,5))

gblur=cv2.GaussianBlur(img1,(5,5),0)


titles=["image","2dconv","blur","gblur"]
imgs=[img1,dst,blur,gblur]
for i in range(4):
    plt.subplot(2,3,i+1),plt.imshow(imgs[i],'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
plt.show()
```

###  median filter 
for removing salt and pepper noise

```python
img=cv2.imread("data/salt_pepper.jpg",1)
```

```python
img1=cv2.medianBlur(img,5)
images=[img,img1]
titles=["image","median blur"]
for i in range(2):
    plt.subplot(1,2,i+1),plt.imshow(images[i],'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
plt.show()
```
