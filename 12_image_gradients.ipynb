```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

<li>compute the gradient (or derivative) of an image</li>
<li>provides information about the intensity change in different directions.</li>
<li> edge detection, feature extraction, and image enhancement.</li>

### Laplacian

```python
img=cv2.imread('data/messi.jpg',cv2.IMREAD_GRAYSCALE)
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
#ksize=5
lap=cv2.Laplacian(img,cv2.CV_64F,ksize=3)
#cv2.CV_64F    data type that support -ve mos that we will be dealing with

lap=np.uint8(np.absolute(lap))

titles=["image","laplacian"]
images=[img,lap]
for i in range(2):
    plt.subplot(1,2,i+1),plt.imshow(images[i])
    plt.title(titles[i])
plt.show()
```

### sobel

```python
lap=cv2.Laplacian(img,cv2.CV_64F,ksize=3)
lap=np.uint8(np.absolute(lap))

#(dx=1,dy=0)order of derivative in x and y direction
sobelx=cv2.Sobel(img,cv2.CV_64F,1,0)
#(dx=0,dy=1)
sobely=cv2.Sobel(img,cv2.CV_64F,0,1)

sobelx=np.uint8(np.absolute(sobelx))
sobely=np.uint8(np.absolute(sobely))

titles=["image","laplacian","sobx","soby"]
images=[img,lap,sobelx,sobely]
for i in range(4):
    plt.subplot(2,2,i+1),plt.imshow(images[i])
    plt.title(titles[i])
plt.show()
```

### canny edge detector

uses a multi stage algorithm to detect edges
<li>gaussian</li>
<li>gradient calculation</li>
<li>non maximum suppression</li>
<li>double threshold</li>
<li>edge track by hysteresis </li>

```python
import cv2
import matplotlib.pyplot as plt

# Read the image in grayscale mode
img = cv2.imread("data/messi.jpg", cv2.IMREAD_GRAYSCALE)

# Apply Canny edge detection
canny = cv2.Canny(img, 100, 200)

imgs = [img, canny]
titles = ["Original", "Canny"]

for i in range(2):
    plt.subplot(1, 2, i+1)
    plt.imshow(imgs[i], cmap='gray')  # Display images in grayscale
    plt.title(titles[i])
    plt.axis('off')  # Hide axis
plt.show()
```
