```python
cap=cv2.VideoCapture(0)
while(True):
    ret,frame=cap.read()
    cv2.imshow('video',frame)
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```

## video capture frame properties

```python
cap=cv2.VideoCapture(0)
while(cap.isOpened()):
    ret,fram=cap.read()
    print(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    print(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    
    gray=cv2.cvtColor(fram,cv2.COLOR_BGR2GRAY)
    cv2.imshow('fram',gray)
    
    if cv2.waitKey(1) & 0XFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
    
    
    
```

```python
cap=cv2.VideoCapture(0)
fourcc=cv2.VideoWriter_fourcc(*'XVID')
out=cv2.VideoWriter('out.avi',fourcc,20.0,(640,480))
while(cap.isOpened()):
    ret,fram=cap.read()
    if ret==True:
        out.write(fram)
        cv2.imshow('fram',fram)
        if cv2.waitKey(1) & 0XFF == ord('q'):
            break
    else:
        break
cap.release()
out.release()
cv2.destroyAllWindows()
    
    
```

```python
import datetime 
```

```python
cap=cv2.VideoCapture(0)
while(cap.isOpened()):
    ret,fram=cap.read()
    font=cv2.FONT_HERSHEY_SIMPLEX
    tex=str(datetime.datetime.now())
    img=cv2.putText(fram,tex,(0,109),font,1,(0,255,255),3,cv2.LINE_AA)
    cv2.imshow('fram',img)
    if cv2.waitKey(1) & 0XFF == ord('q'):
            break
    
cap.release()
cv2.destroyAllWindows()
```
